```python
# Machine Learning Model Performance Analysis
# Google Colab Compatible Version

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)

print("🚀 Starting Machine Learning Model Performance Analysis")
print("=" * 60)

# Generate synthetic data to simulate the original analysis
def generate_synthetic_data(model_name, n_samples=8012):
    """
    Generate synthetic experimental and predicted release data
    to simulate the behavior of different ML models
    """
    # Create time points
    time = np.random.uniform(0, 100, n_samples)

    # Generate experimental release values (ground truth)
    experimental_release = 0.3 * np.sin(time/10) + 0.5 * np.exp(-time/20) + 0.2 * np.random.normal(0, 0.1, n_samples)
    experimental_release = np.clip(experimental_release, 0, 1)  # Keep values between 0 and 1

    # Model-specific prediction patterns (simulating different model performances)
    model_params = {
        'LGBM': {'noise': 0.05, 'bias': 0.02},
        'RF': {'noise': 0.06, 'bias': 0.03},
        'NGB': {'noise': 0.07, 'bias': 0.04},
        'DT': {'noise': 0.12, 'bias': 0.08},
        'kNN': {'noise': 0.11, 'bias': 0.09},
        'SVR': {'noise': 0.15, 'bias': 0.12},
        'lasso': {'noise': 0.13, 'bias': 0.11},
        'PLS': {'noise': 0.16, 'bias': 0.13},
        'MLR': {'noise': 0.17, 'bias': 0.14}
    }

    params = model_params.get(model_name, {'noise': 0.1, 'bias': 0.1})

    # Generate predicted values with model-specific characteristics
    predicted_release = experimental_release + np.random.normal(params['bias'], params['noise'], n_samples)
    predicted_release = np.clip(predicted_release, 0, 1)

    # Create experimental indices
    experimental_index = np.arange(n_samples)

    # Calculate absolute error
    absolute_error = np.abs(experimental_release - predicted_release)

    # Create DataFrame
    df = pd.DataFrame({
        'Time': time,
        'Experimental_Release': experimental_release,
        'Predicted_Release': predicted_release,
        'Experimental Index': experimental_index,
        'Absolute_Error': absolute_error
    })

    return df

# Generate data for all models
print("📊 Generating synthetic data for all machine learning models...")
model_names = ['LGBM', 'RF', 'NGB', 'DT', 'kNN', 'SVR', 'lasso', 'PLS', 'MLR']

model_data = {}
for model in model_names:
    model_data[f'ALL_{model}'] = generate_synthetic_data(model)
    print(f"✅ Generated data for {model}")

print("\n🎯 Data Generation Complete!")
print("=" * 60)

# Display basic statistics for all models
print("\n📈 MODEL PERFORMANCE STATISTICS")
print("=" * 60)

# Create a summary DataFrame
summary_stats = []
for model in model_names:
    df = model_data[f'ALL_{model}']
    stats = {
        'Model': model,
        'Mean_AE': df['Absolute_Error'].mean(),
        'Std_AE': df['Absolute_Error'].std(),
        'Min_AE': df['Absolute_Error'].min(),
        'Max_AE': df['Absolute_Error'].max(),
        'Median_AE': df['Absolute_Error'].median(),
        'R2_Score': r2_score(df['Experimental_Release'], df['Predicted_Release'])
    }
    summary_stats.append(stats)

summary_df = pd.DataFrame(summary_stats)
summary_df = summary_df.sort_values('Mean_AE')
print(summary_df.round(6))

# Create visualizations
print("\n🎨 Creating Visualizations...")
print("=" * 60)

# 1. Box Plot of Absolute Errors
plt.figure(figsize=(15, 8))
plt.subplot(2, 2, 1)

# Prepare data for box plot
box_data = []
box_labels = []
for model in model_names:
    box_data.append(model_data[f'ALL_{model}']['Absolute_Error'])
    box_labels.append(model)

plt.boxplot(box_data, labels=box_labels)
plt.xticks(rotation=45)
plt.title('Absolute Error Distribution by Model')
plt.ylabel('Absolute Error')
plt.grid(True, alpha=0.3)

# 2. Mean Absolute Error Comparison
plt.subplot(2, 2, 2)
mean_errors = [model_data[f'ALL_{model}']['Absolute_Error'].mean() for model in model_names]
colors = plt.cm.viridis(np.linspace(0, 1, len(model_names)))
bars = plt.bar(model_names, mean_errors, color=colors)
plt.title('Mean Absolute Error by Model')
plt.ylabel('Mean Absolute Error')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Add value labels on bars
for bar, error in zip(bars, mean_errors):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.005,
             f'{error:.3f}', ha='center', va='bottom', fontsize=8)

# 3. Predicted vs Experimental (for best model)
plt.subplot(2, 2, 3)
best_model_idx = np.argmin(mean_errors)
best_model = model_names[best_model_idx]
best_data = model_data[f'ALL_{best_model}']

plt.scatter(best_data['Experimental_Release'], best_data['Predicted_Release'],
           alpha=0.5, s=1)
plt.plot([0, 1], [0, 1], 'r--', label='Perfect Prediction')
plt.xlabel('Experimental Release')
plt.ylabel('Predicted Release')
plt.title(f'Predicted vs Experimental ({best_model} - Best Model)')
plt.legend()
plt.grid(True, alpha=0.3)

# 4. R² Score Comparison
plt.subplot(2, 2, 4)
r2_scores = [r2_score(model_data[f'ALL_{model}']['Experimental_Release'],
                     model_data[f'ALL_{model}']['Predicted_Release']) for model in model_names]
bars = plt.bar(model_names, r2_scores, color=colors)
plt.title('R² Score by Model')
plt.ylabel('R² Score')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# Add value labels on bars
for bar, r2 in zip(bars, r2_scores):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01,
             f'{r2:.3f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()

# Additional Analysis: Statistical Tests and Detailed Metrics
print("\n📊 DETAILED PERFORMANCE METRICS")
print("=" * 60)

detailed_metrics = []
for model in model_names:
    df = model_data[f'ALL_{model}']
    exp = df['Experimental_Release']
    pred = df['Predicted_Release']

    metrics = {
        'Model': model,
        'MAE': mean_absolute_error(exp, pred),
        'MSE': mean_squared_error(exp, pred),
        'RMSE': np.sqrt(mean_squared_error(exp, pred)),
        'R²': r2_score(exp, pred),
        'Count': len(df),
        'AE_Q25': df['Absolute_Error'].quantile(0.25),
        'AE_Q75': df['Absolute_Error'].quantile(0.75)
    }
    detailed_metrics.append(metrics)

detailed_df = pd.DataFrame(detailed_metrics)
detailed_df = detailed_df.sort_values('MAE')
print(detailed_df.round(6))

# Create a heatmap of model performance
print("\n🔥 Creating Performance Heatmap...")
plt.figure(figsize=(12, 8))

# Normalize metrics for heatmap (lower is better for error metrics)
heatmap_data = detailed_df[['MAE', 'MSE', 'RMSE']].values
heatmap_data_normalized = (heatmap_data - heatmap_data.min(axis=0)) / (heatmap_data.max(axis=0) - heatmap_data.min(axis=0))

# For R², higher is better, so we invert it
r2_normalized = 1 - ((detailed_df['R²'].values - detailed_df['R²'].min()) / (detailed_df['R²'].max() - detailed_df['R²'].min()))

# Combine all metrics
combined_metrics = np.column_stack([heatmap_data_normalized, r2_normalized.reshape(-1, 1)])
metric_names = ['MAE', 'MSE', 'RMSE', 'R² (inverted)']

sns.heatmap(combined_metrics.T,
            xticklabels=detailed_df['Model'],
            yticklabels=metric_names,
            annot=True,
            cmap='RdYlBu_r',
            fmt='.3f',
            cbar_kws={'label': 'Normalized Performance (0=Best, 1=Worst)'})
plt.title('Model Performance Heatmap\n(Darker = Better Performance)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Final Summary
print("\n🏆 FINAL SUMMARY")
print("=" * 60)
print(f"🥇 Best Model (Lowest MAE): {detailed_df.iloc[0]['Model']} (MAE: {detailed_df.iloc[0]['MAE']:.6f})")
print(f"🥈 Second Best: {detailed_df.iloc[1]['Model']} (MAE: {detailed_df.iloc[1]['MAE']:.6f})")
print(f"🥉 Third Best: {detailed_df.iloc[2]['Model']} (MAE: {detailed_df.iloc[2]['MAE']:.6f})")
print(f"\n📉 Worst Model: {detailed_df.iloc[-1]['Model']} (MAE: {detailed_df.iloc[-1]['MAE']:.6f})")

print(f"\n📊 Dataset Information:")
print(f"   • Total Samples per Model: {detailed_df.iloc[0]['Count']:,}")
print(f"   • Models Analyzed: {len(model_names)}")
print(f"   • Best R² Score: {detailed_df['R²'].max():.6f} ({detailed_df.loc[detailed_df['R²'].idxmax(), 'Model']})")

print("\n✨ Analysis Complete! All visualizations have been generated.")
print("🔄 You can run individual sections or modify parameters as needed.")
```
